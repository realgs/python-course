```python
from scipy import linspace , cos , exp, random, meshgrid, zeros
from scipy.optimize import fmin
from matplotlib.pyplot import plot, show, legend, figure, cm, contour, clabel

def f(x):
    return cos(x) - 3 * exp ( -(x - 0.2) ** 2)

x0 = 6*random.rand()
x_min = fmin (f, x0)

delta = 3
x = linspace(x_min - delta, x_min + delta, 301)
y = f(x)
plot(x,y)
plot(x0, f(x0), 'go', label='initial x')
plot(x_min, f(x_min), 'ro', label='optimal x')
legend(loc='upper right')
show()
```

```python
% matplotlib notebook

def f(x):
    return (x[0]-1)**2 + (x[1] + 2)**2

x0 = random.randn(2)
x_min = fmin(f, x0)

from mpl_toolkits.mplot3d import Axes3D

delta = 7
x_knots = linspace(x_min[0]-delta, x_min[0]+delta, 41)
y_knots = linspace(x_min[1]-delta, x_min[1]+delta, 41)
X, Y = meshgrid(x_knots, y_knots)
Z = zeros(X.shape)
for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
        Z[i][j] = f([X[i,j],Y[i,j]])

ax = Axes3D(figure(figsize=(8,5)))
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.4)
ax.plot([x0[0]], [x0[1]], [f(x0)], color='g',marker='o', markersize=5, label='initial')
ax.plot([x_min[0]], [x_min[1]], [f(x_min)], color='k',marker='o', markersize=5, label='final')
ax.legend()
show()
```

```python
from scipy import linspace , cos , exp, random, meshgrid, zeros, sin
from scipy.optimize import fmin
from matplotlib.pyplot import plot, show, legend, figure, cm, contour, clabel

% matplotlib notebook

def f(w):
    x = w[0]
    y = w[1]
    return -(10*sin(x+y) + (x-y)**2 - 1.5*x + 2.5*y + 1)

def g(w):
    return -f(w)

x0 = random.rand(2)*20
x_min = fmin(g, x0)

from mpl_toolkits.mplot3d import Axes3D

delta = 7
x_knots = linspace(x_min[0]-delta, x_min[0]+delta, 41)
y_knots = linspace(x_min[1]-delta, x_min[1]+delta, 41)
X, Y = meshgrid(x_knots, y_knots)
Z = zeros(X.shape)
for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
        Z[i][j] = f([X[i,j],Y[i,j]])

ax = Axes3D(figure(figsize=(8,5)))
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.4)
ax.plot([x0[0]], [x0[1]], [f(x0)], color='g',marker='o', markersize=5, label='initial')
ax.plot([x_min[0]], [x_min[1]], [f(x_min)], color='k',marker='o', markersize=5, label='final')
ax.legend()
show()
```

```python
from scipy import linspace , cos , exp, random, meshgrid, zeros, sin
from scipy.optimize import fmin
from matplotlib.pyplot import plot, show, legend, figure, cm, contour, clabel

% matplotlib notebook

def f(w):
    x = w[0]
    y = w[1]
    return 2*x**2 + 1*y**2 + (0.5*x*y)

def g(w):
    return -f(w)

x0 = random.rand(2)
x_min = fmin(f, x0)

from mpl_toolkits.mplot3d import Axes3D

delta = 2
x_knots = linspace(x_min[0]-delta, x_min[0]+delta, 41)
y_knots = linspace(x_min[1]-delta, x_min[1]+delta, 41)
X, Y = meshgrid(x_knots, y_knots)
Z = zeros(X.shape)
for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
        Z[i][j] = f([X[i,j],Y[i,j]])

ax = Axes3D(figure(figsize=(8,5)))
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.4)
ax.plot([x0[0]], [x0[1]], [f(x0)], color='g',marker='o', markersize=5, label='initial')
ax.plot([x_min[0]], [x_min[1]], [f(x_min)], color='k',marker='o', markersize=5, label='final')
ax.legend()
show()
```
