```python
from numpy import *
from matplotlib.pyplot import *

plot([1,2,3,2,3,4,3,4,5])
show()
```

```python
x = linspace(0, 4.*pi, 100)
y = sin(x)
plot(x, y)
show()
```

```python
x = linspace(-10., 10., 200)
x2 = linspace(-10., 10., 200)
y = sin(x) * exp(-(x/5.0)**2)
y2 = cos(x2) * exp(-(x2/2.0)**2)
figure(1, figsize = (6,4) )
plot(x, y, 'b-', label='theory')
plot(x2, y2, 'ro', label="data")
xlabel('x')
legend(loc='upper right')
axhline(color = 'gray', zorder=-1)
axvline(color = 'gray', zorder=-1)
ylim(-1.3, 1.3)
savefig('WavyPulse.pdf')
show()
```

```python
from numpy.random import *

t = linspace(0, 1, 101)
subplot(2, 2, 1); hist(rand(20))
subplot(2, 2, 2); plot(t, t**2, t, t**3 - t)
subplot(2, 2, 3); plot(rand(20), rand(20), 'r*')
subplot(2, 2, 4); plot(t*cos(10*t), t*sin(10*t))
show()
```

```python
X = linspace(0,4*pi, 501)
Y1 = sin(X)
Y2 = sin(2*X)
figure(1, figsize=(13,5))
subplot(1,2,1)
plot(X,Y1,X,Y2)
subplot(1,2,2)
plot(Y1,Y2)
show()
```

```python
X = linspace(0,20, 1001)
Y = zeros(X.shape)

N = 100
for i in range(1,N+1,2):
    Y += np.sin(i*X)/i

plot(X,Y)
show()
```

```python
# na listach
num_points = 20000
x_values = [0]
y_values = [0]

while len(x_values) < num_points:  
    
    next_x = x_values[-1] + randn()
    next_y = y_values[-1] + randn()
    
    x_values.append(next_x)
    y_values.append(next_y)

point_numbers = range(num_points)
scatter(x_values, y_values, s=1, c=point_numbers, cmap=cm.Blues)
# pierwszy i ostatni punkt
scatter(x_values[0], y_values[0], c='green', edgecolors='none', s=70)
scatter(x_values[-1], y_values[-1], c='red', edgecolors='none', s=70)
# ukrycie osi
axes().get_xaxis().set_visible(False)
axes().get_yaxis().set_visible(False)
show()
figure(dpi=128, figsize=(7,4))
```

```python
# na tablicach
num_points = 20000
x_values = zeros(num_points)
y_values = zeros(num_points)

for i in range(num_points-1):  
    
    x_values[i+1] = x_values[i] + randn()
    y_values[i+1] = y_values[i] + randn()


point_numbers = range(num_points)
scatter(x_values, y_values, s=1, c=point_numbers, cmap=cm.Blues)
# pierwszy i ostatni punkt
scatter(x_values[0], y_values[0], c='green', edgecolors='none', s=70)
scatter(x_values[-1], y_values[-1], c='red', edgecolors='none', s=70)
# ukrycie osi
axes().get_xaxis().set_visible(False)
axes().get_yaxis().set_visible(False)
show()
figure(dpi=128, figsize=(7,4))
```
